````python
import matplotlib.pyplot as plt
import seaborn as sns
````

# Graficos Categ√≥ricos

````python
tips = sns.load_dataset('tips')
tips.head(2)
````

## Countplot

````python
sns.countplot(data=tips, x='day', hue='sex');
````

## Stripplot

````python
plt.figure(figsize=(10,6))
sns.stripplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True);
````

## Swarmplot

````python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True)
plt.show()
````

## Boxplot
Se puede graficar tanto variables categoricas(imagen 1) como numericas(imagen 2)

````python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True)
sns.swarmplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True, color='0', marker='<')
plt.show() 
````

````python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, y='total_bill')
plt.show() 
````

## Violinplot
Al igual que el boxplot, podemos graficar variables categoricas(imagen 1 y 2) o numericas(imagen 3)

````python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day' , y='total_bill', hue='sex' ,palette='pastel')
plt.show() 
````

````python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,x='day' , y='total_bill', hue='sex', split=True,palette='pastel')
plt.show() 
````

````python
plt.figure(figsize=(6,6))
sns.violinplot(data=tips,y='total_bill')
plt.show() 
````

## Catplot
Similar a ```displot```, pero sirve para graficar graficos categoricos

````python
sns.catplot(data=tips,x='day' , y='total_bill', hue='sex')
plt.show()
````

## Catplot
Similar a ```displot```, pero sirve para graficar graficos categoricos

````python
sns.catplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True)
plt.show()
````

````python
sns.catplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True, kind='swarm')
plt.show()
````

````python
sns.catplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True, kind='box')
plt.show()
````

````python
sns.catplot(data=tips,x='day' , y='total_bill', hue='sex', dodge=True, kind='boxen', col='time', orient='v')
plt.show()
````
