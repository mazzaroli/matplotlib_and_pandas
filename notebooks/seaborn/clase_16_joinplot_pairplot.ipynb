```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

# Joinplot & Pairplot
[Documentacion Joinplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

[Documentacion Pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)


```python
sns.jointplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25, fill=True, multiple='dodge'))
plt.show()
```

# Pairplot

```python
sns.pairplot(data=tips)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex')
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True)
plt.show()
```

```python
sns.pairplot(data=tips, hue='sex',corner=True,kind='scatter')
plt.show()
```

```python
sns.pairplot(data=tips, hue='total_bill', kind='scatter')
```

```python
sns.pairplot(data=tips, hue='tip', kind='reg')
```
