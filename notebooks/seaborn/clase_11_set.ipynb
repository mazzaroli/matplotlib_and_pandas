```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Set

```python
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

[Documentacion Paleta de Colores](https://seaborn.pydata.org/generated/seaborn.color_palette.html#seaborn.color_palette)

[Documentacion Set Temas](https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme)

```python
sns.set(
    style='ticks', 
    palette='dark', 
    font='Verdana',
    font_scale=1,
    )

sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(
    style='ticks',
    palette='husl',
    font='Verdana',
    font_scale=2,
    )

sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```
