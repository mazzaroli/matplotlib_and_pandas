```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

# Heatmap

### pandas.corr()

Funcion de pandas para poder ver la correlacion de las variables

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
plt.show()
```

```python
sns.heatmap(data=(tips.corr()), annot=(True))
plt.show()
```

```python
sns.heatmap(data=(tips.corr()), annot=(True), cmap='RdYlGn')
plt.show()
```

```python
sns.heatmap(data=(tips.corr()), annot=(True), cmap='RdYlGn', linewidths=5)
plt.show()
```

```python
sns.heatmap(data=(tips.corr()), annot=(True), cmap='coolwarm', linewidths=5, linecolor='black')
plt.show()
```

```python
sns.heatmap(data=(tips.corr()), annot=(True), cmap='coolwarm', linewidths=5, linecolor='black', vmin=0, vmax=1)
plt.show()
```

```python
sns.heatmap(data=(tips.corr()), annot=(True), cmap='coolwarm', linewidths=5, linecolor='black', vmin=0, vmax=1, cbar=False)
plt.show()
```

```python
sns.heatmap(data=(tips.corr()), annot=(True), cmap='coolwarm', linewidths=5, linecolor='black', vmin=0, vmax=1, cbar=True)
plt.show()
```
