```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

# Graficos de Relacion

```python
plt.figure(figsize=(8,8))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day',style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12, 0.5))
plt.show()
```

```python
markers = {'Lunch':'D', 'Dinner':'s'}

plt.figure(figsize=(8,8))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day',style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.1 , 0.5))
plt.show()
```

## lineplot

```python
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day',style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.1,0.5))
plt.show()
```

## relplot

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day',style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.1,0.5))
plt.show()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day',style='time', size='size', col='size')
# plt.legend(loc='center', bbox_to_anchor=(1.1,0.5))
plt.show()
```
