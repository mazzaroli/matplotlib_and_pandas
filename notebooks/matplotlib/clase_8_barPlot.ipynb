```python
import numpy as np
import matplotlib.pyplot as plt
```

# Bar Plot

El grafico de barras es algo que nos permite graficar variables categoricas, osea variables de texto

```python
country = ['INDIA','JAPON','MEXICO','COLOMBIA','GERMANY',]
population = [1000,800,900,1000,300]
```

```python
plt.style.use('dark_background')
plt.bar(
    country,
    population, 
    width=0.5, 
    color=['#9E9186','#BCEBBD','#EBE4DF','#CEC7EB','#918E9E']
    )

plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'), rotation=0)

plt.ylabel('POBLACION por Miles')
plt.title('POBLACION POR PAIS\n(Valores no reales)')

plt.show()
```

# Locuras con Pandas 

```python
import pandas as pd
```

```python
dictionary = {
    'country': ['INDIA','JAPON','MEXICO','COLOMBIA','GERMANY'],
    'population': [1000,800,900,1000,300]
}
```

```python
df_population_by_country = pd.DataFrame(dictionary)
df_population_by_country
```

```python
plt.style.use('tableau-colorblind10')

fig, bar = plt.subplots(1,1, figsize=(5,5))

plt.bar(
    data=df_population_by_country, 
    x='country', 
    height='population',
    color=['#9E9186','#BCEBBD','#EBE4DF','#CEC7EB','#918E9E'],
    width=0.5
)

plt.title('')
```
