```python
import numpy as np
import matplotlib.pyplot as plt
```

# Subplot
Subplot permite crear graficos dentro de un grafico. Esto lo hace a traves de una matriz de graficos y se puede acceder a ellos a traves de los indices


Estructura:
*subplot(nrows, ncols, index)*


```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b--')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b--')

plt.subplot(2,1,2) # Grafico 2
plt.pie(y)
plt.show()
```
